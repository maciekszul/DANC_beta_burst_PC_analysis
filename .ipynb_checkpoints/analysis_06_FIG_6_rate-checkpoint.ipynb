```python
import pickle
import numpy as np
import pandas as pd
import os.path as op
import itertools as it
from mne import read_epochs
import matplotlib.pylab as plt
from tqdm.notebook import trange, tqdm
```

```python
visual_epoch_file = "/home/mszul/datasets/explicit_implicit_beta/derivatives/processed/sub-145/sub-145-002-visual-epo.fif"
motor_epoch_file = "/home/mszul/datasets/explicit_implicit_beta/derivatives/processed/sub-145/sub-145-002-motor-epo.fif"
burst_features_file = "/home/mszul/datasets/explicit_implicit_beta/derivatives/PCA_results/burst_features.csv"
waveform_array_file = "/home/mszul/datasets/explicit_implicit_beta/derivatives/PCA_results/all_waveforms.npy"
```

```python
visual_epoch = read_epochs(visual_epoch_file, verbose=False)
motor_epoch = read_epochs(motor_epoch_file, verbose=False)
visual_epoch_times = visual_epoch.times
motor_epoch_times = motor_epoch.times
del visual_epoch
del motor_epoch
```

```python
burst_rate_overall_file = "/home/mszul/datasets/explicit_implicit_beta/derivatives/PCA_results/burst_rate_overall.pkl"
burst_rate_overall = pickle.load(open(burst_rate_overall_file, "rb"))
brate_mean_vis = np.mean(burst_rate_overall["vis"], axis=0)
brate_sem_vis = np.std(burst_rate_overall["vis"], axis=0) / np.sqrt(36)
brate_mean_mot = np.mean(burst_rate_overall["mot"], axis=0)
brate_sem_mot = np.std(burst_rate_overall["mot"], axis=0) / np.sqrt(36)
```

```python
buffer = 0.125
bin_width=0.05
baseline_range = [-0.5, -0.25]

visual_time_bins = np.arange(visual_epoch_times[0] + buffer, visual_epoch_times[-1] - buffer, bin_width)
motor_time_bins = np.arange(motor_epoch_times[0] + buffer, motor_epoch_times[-1] - buffer, bin_width)

visual_bin_ranges = list(zip(visual_time_bins[:-1], visual_time_bins[1:]))
motor_bin_ranges = list(zip(motor_time_bins[:-1], motor_time_bins[1:]))
vis_time_plot = visual_time_bins[:-1]
mot_time_plot = motor_time_bins[:-1]
vis_time_ix = np.where(
    (visual_epoch_times >= vis_time_plot[0]) &
    (visual_epoch_times <= vis_time_plot[-1])
)[0]
mot_time_ix = np.where(
    (motor_epoch_times >= mot_time_plot[0]) &
    (motor_epoch_times <= mot_time_plot[-1])
)[0]

bin_ranges = {
    "vis": visual_time_bins,
    "mot": motor_time_bins
}

baseline_ixs = np.where(
    (visual_time_bins >= baseline_range[0]) &
    (visual_time_bins <= baseline_range[-1])
)
```

```python
vis_time_plot
```

```python
burst_features = pd.read_csv(burst_features_file)
waveform_array_file = "/home/mszul/datasets/explicit_implicit_beta/derivatives/PCA_results/all_waveforms.npy"
waveform_array = np.load(waveform_array_file)
waveform_time = np.linspace(-.13, .13, num=156)
mean_waveform = np.mean(waveform_array, axis=0)
```

```python
burst_features.columns
```

```python
burst_features.index.to_numpy()
```

```python
prct = np.linspace(0,100, num=5)
prct_ranges = list(zip(prct[:-1], prct[1:]))

quartile_ixes = {}
wvfrm = []
for ix, (b, e) in enumerate(prct_ranges):
    key = ix
    low_ = np.percentile(burst_features.peak_amp_base, b)
    upp_ = np.percentile(burst_features.peak_amp_base, e)
    ixs = burst_features.loc[
        (burst_features.peak_amp_base >= low_) &
        (burst_features.peak_amp_base <= upp_)
    ].index.to_numpy()
    quartile_ixes[key] = low_, upp_
    wvfrm.append(np.mean(waveform_array[ixs, :], axis=0))
```

```python
subjects = burst_features.subject.unique()
epochs = ["vis", "mot"]
results = {
    i: {
        c: [] for c in range(4)
    } for i in epochs
}
for qrt_ix,(b, e) in tqdm(enumerate(prct_ranges)):
    for sub in tqdm(subjects):
        subset = burst_features.loc[
            (burst_features.subject == sub)
        ]
        sub_res = {i: [] for i in epochs}
        for epoch in epochs:
            trial_res = []
            for tr_ix in subset.pp_ix.unique():
                low_, upp_ = quartile_ixes[qrt_ix]
                subset_tr = subset.loc[
                    (subset.epoch == epoch) &
                    (subset.pp_ix == tr_ix) &
                    (subset.peak_amp_base >= low_) &
                    (subset.peak_amp_base <= upp_)
                ]
                hist, bins = np.histogram(subset_tr.peak_time, bin_ranges[epoch])
                trial_res.append(hist)
            sub_res[epoch] = np.nanmean(trial_res, axis=0)
        baseline = np.mean(sub_res["vis"][baseline_ixs])
        sub_res["vis"] = (sub_res["vis"] - baseline) / baseline
        sub_res["mot"] = (sub_res["mot"] - baseline) / baseline
        results["vis"][qrt_ix].append(sub_res["vis"])
        results["mot"][qrt_ix].append(sub_res["mot"])
```

```python
subjects = burst_features.subject.unique()
epochs = ["vis", "mot"]
results_ova = {
    i: [] for i in epochs
}

for sub in tqdm(subjects):
    subset = burst_features.loc[
        (burst_features.subject == sub)
    ]
    sub_res = {i: [] for i in epochs}
    for epoch in epochs:
        trial_res = []
        for tr_ix in subset.pp_ix.unique():
            subset_tr = subset.loc[
                (subset.epoch == epoch) &
                (subset.pp_ix == tr_ix)
            ]
            hist, bins = np.histogram(subset_tr.peak_time, bin_ranges[epoch])
            trial_res.append(hist)
        sub_res[epoch] = np.nanmean(trial_res, axis=0)
    baseline = np.mean(sub_res["vis"][baseline_ixs])
    sub_res["vis"] = (sub_res["vis"] - baseline) / baseline
    sub_res["mot"] = (sub_res["mot"] - baseline) / baseline
    results_ova["vis"].append(sub_res["vis"])
    results_ova["mot"].append(sub_res["mot"])
```

```python
col_r = plt.cm.gist_rainbow(np.linspace(0,1, num=4))
labels=["1st", "2nd", "3rd", "4th"]
f, ax = plt.subplots(1,3, figsize=(12,3), constrained_layout=True, gridspec_kw={"width_ratios":[3, 5, 4]}, dpi=300)
for ix, wf in enumerate(wvfrm):
    ax[0].plot(waveform_time, wf/1e-15, lw=1, color=col_r[ix], label=labels[ix]+" QRT")
ax[0].plot(waveform_time, mean_waveform/1e-15, lw=1, color="black", label="Mean")
ax[0].legend(fontsize="small")


for qrt_ix in range(4):
    vis = np.vstack(results["vis"][qrt_ix])
    vis_mean = np.mean(vis, axis=0)
    vis_sem = np.std(vis, axis=0)/np.sqrt(36)
    mot = np.vstack(results["mot"][qrt_ix])
    mot_mean = np.mean(mot, axis=0)
    mot_sem = np.std(mot, axis=0)/np.sqrt(36)
    
    ax[1].plot(vis_time_plot, vis_mean, c=col_r[qrt_ix])
    ax[1].fill_between(
        vis_time_plot,
        vis_mean - vis_sem,
        vis_mean + vis_sem,
        color=col_r[qrt_ix],
        lw=0,
        alpha=0.2
    )
    ax[2].plot(mot_time_plot, mot_mean, c=col_r[qrt_ix])
    ax[2].fill_between(
        mot_time_plot,
        mot_mean - mot_sem,
        mot_mean + mot_sem,
        color=col_r[qrt_ix],
        lw=0,
        alpha=0.2,
        label=labels[qrt_ix]
    )

ax[1].plot(vis_time_plot, brate_mean_vis, c="black", linestyle=(0, (5,10)))
ax[1].fill_between(
    vis_time_plot,
    brate_mean_vis - brate_sem_vis,
    brate_mean_vis + brate_sem_vis,
    color="black",
    lw=0,
    alpha=0.2
)
ax[2].plot(mot_time_plot, brate_mean_mot, c="black", linestyle=(0, (5,10)))
ax[2].fill_between(
    mot_time_plot,
    brate_mean_mot - brate_sem_mot,
    brate_mean_mot + brate_sem_mot,
    color="black",
    lw=0,
    alpha=0.2,
    label="Overall\nBurst rate"
)

ax[2].legend(fontsize="small")
```

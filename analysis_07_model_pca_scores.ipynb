```python
import numpy as np
import matplotlib.pyplot as plt
import hnn_core
import copy
import pickle as pk
from scipy.interpolate import interp1d
from scipy.stats import spearmanr
from sklearn.preprocessing import RobustScaler
from hnn_core import Dipole, read_params, Network, JoblibBackend, simulate_dipole
```

```python
basic_dpls=np.load('basic_bursts.npy')

prox_t0_range=np.linspace(-25,25,10)
prox_shift_dpls=np.load('prox_shift_bursts.npy')
dist_ampa_range=np.linspace(4.5e-5,8.5e-5,10)
dist_strength_dpls=np.load('dist_strength_bursts.npy')
prox_ampa_range=np.linspace(0.5e-5,5.5e-5,10)
prox_strength_dpls=np.load('prox_strength_bursts.npy')
dist_width_range=np.linspace(1,15,10)
dist_width_dpls=np.load('dist_width_bursts.npy')
prox_width_range=np.linspace(10,40,10)
prox_width_dpls=np.load('prox_width_bursts.npy')

dt=40000
sim_times=np.linspace(-100,100.,int(.200*dt)+1)

data_times=np.linspace(-130,130,157)
data_times=data_times[:-1]
all_bursts=np.load('all_waveforms.npy')

rs_dpls=np.zeros((basic_dpls.shape[0],len(data_times)))
for i in range(basic_dpls.shape[0]):
    f = interp1d(sim_times, basic_dpls[i,:], kind='linear', fill_value="extrapolate")
    rs_dpls[i,:]=f(data_times)
```

```python
scale_factor=9e-16
scaled_dpls=rs_dpls*scale_factor
scaler = RobustScaler()
burst_std=scaler.fit_transform(rs_dpls)
```

```python
plt.plot(data_times,np.mean(scaled_dpls,axis=0))
plt.plot(data_times,np.mean(all_bursts,axis=0))
```

```python
pca = pk.load(open("pca_solution_MEG.pkl",'rb'))
scaler = pk.load(open("scaler_MEG.pkl",'rb'))

data_bursts_std=scaler.transform(all_bursts)
data_pcs=pca.transform(data_bursts_std)

rs_dpls=np.zeros((basic_dpls.shape[0],len(data_times)))
for i in range(basic_dpls.shape[0]):
    f = interp1d(sim_times, basic_dpls[i,:], kind='cubic', fill_value="extrapolate")
    rs_dpls[i,:]=f(data_times)*scale_factor
burst_std=scaler.transform(rs_dpls)

# Convert to femtoTesla
base_sim_pcs = pca.transform(burst_std)

rs_dpls=np.zeros((prox_shift_dpls.shape[0],prox_shift_dpls.shape[1],len(data_times)))
prox_shift_pcs=np.zeros((prox_shift_dpls.shape[0],prox_shift_dpls.shape[1],20))
for i in range(prox_shift_dpls.shape[0]):
    for j in range(prox_shift_dpls.shape[1]):
        f = interp1d(sim_times, prox_shift_dpls[i,j,:], kind='cubic', fill_value="extrapolate")
        rs_dpls[i,j,:]=f(data_times)*scale_factor
    rs_dpls[i,:,:]=scaler.transform(rs_dpls[i,:,:])
    prox_shift_pcs[i,:,:]=pca.transform(rs_dpls[i,:,:])
    
rs_dpls=np.zeros((dist_strength_dpls.shape[0],dist_strength_dpls.shape[1],len(data_times)))
dist_strength_pcs=np.zeros((dist_strength_dpls.shape[0],dist_strength_dpls.shape[1],20))
for i in range(dist_strength_dpls.shape[0]):
    for j in range(dist_strength_dpls.shape[1]):
        f = interp1d(sim_times, dist_strength_dpls[i,j,:], kind='cubic', fill_value="extrapolate")
        rs_dpls[i,j,:]=f(data_times)*scale_factor
    rs_dpls[i,:,:]=scaler.transform(rs_dpls[i,:,:])
    dist_strength_pcs[i,:,:]=pca.transform(rs_dpls[i,:,:])
    
rs_dpls=np.zeros((prox_strength_dpls.shape[0],prox_strength_dpls.shape[1],len(data_times)))
prox_strength_pcs=np.zeros((prox_strength_dpls.shape[0],prox_strength_dpls.shape[1],20))
for i in range(prox_strength_dpls.shape[0]):
    for j in range(prox_strength_dpls.shape[1]):
        f = interp1d(sim_times, prox_strength_dpls[i,j,:], kind='cubic', fill_value="extrapolate")
        rs_dpls[i,j,:]=f(data_times)*scale_factor
    rs_dpls[i,:,:]=scaler.transform(rs_dpls[i,:,:])
    prox_strength_pcs[i,:,:]=pca.transform(rs_dpls[i,:,:])
    
rs_dpls=np.zeros((dist_width_dpls.shape[0],dist_width_dpls.shape[1],len(data_times)))
dist_width_pcs=np.zeros((dist_width_dpls.shape[0],dist_width_dpls.shape[1],20))
for i in range(dist_width_dpls.shape[0]):
    for j in range(dist_width_dpls.shape[1]):
        f = interp1d(sim_times, dist_width_dpls[i,j,:], kind='cubic', fill_value="extrapolate")
        rs_dpls[i,j,:]=f(data_times)*scale_factor
    rs_dpls[i,:,:]=scaler.transform(rs_dpls[i,:,:])
    dist_width_pcs[i,:,:]=pca.transform(rs_dpls[i,:,:])
    
rs_dpls=np.zeros((prox_width_dpls.shape[0],prox_width_dpls.shape[1],len(data_times)))
prox_width_pcs=np.zeros((prox_width_dpls.shape[0],prox_width_dpls.shape[1],20))
for i in range(prox_width_dpls.shape[0]):
    for j in range(prox_width_dpls.shape[1]):
        f = interp1d(sim_times, prox_width_dpls[i,j,:], kind='cubic', fill_value="extrapolate")
        rs_dpls[i,j,:]=f(data_times)*scale_factor
    rs_dpls[i,:,:]=scaler.transform(rs_dpls[i,:,:])
    prox_width_pcs[i,:,:]=pca.transform(rs_dpls[i,:,:])
    
```

```python
prox_t0=np.zeros(len(prox_t0_range)*prox_shift_pcs.shape[1])
prox_shift_score=np.zeros((len(prox_t0_range)*prox_shift_pcs.shape[1],18))
for i in range(len(prox_t0_range)):
    for j in range(prox_shift_pcs.shape[1]):
        prox_t0[i*prox_shift_pcs.shape[1]+j]=prox_t0_range[i]
        prox_shift_score[i*prox_shift_pcs.shape[1]+j,:]=prox_shift_pcs[i,j,:18]

dist_ampa=np.zeros(len(dist_ampa_range)*dist_strength_pcs.shape[1])
dist_strength_score=np.zeros((len(dist_ampa_range)*dist_strength_pcs.shape[1],18))
for i in range(len(dist_ampa_range)):
    for j in range(dist_strength_pcs.shape[1]):
        dist_ampa[i*dist_strength_pcs.shape[1]+j]=dist_ampa_range[i]
        dist_strength_score[i*dist_strength_pcs.shape[1]+j,:]=dist_strength_pcs[i,j,:18]

prox_ampa=np.zeros(len(prox_ampa_range)*prox_strength_pcs.shape[1])
prox_strength_score=np.zeros((len(prox_ampa_range)*prox_strength_pcs.shape[1],18))
for i in range(len(prox_ampa_range)):
    for j in range(prox_strength_pcs.shape[1]):
        prox_ampa[i*prox_strength_pcs.shape[1]+j]=prox_ampa_range[i]
        prox_strength_score[i*prox_strength_pcs.shape[1]+j,:]=prox_strength_pcs[i,j,:18]

dist_width=np.zeros(len(dist_width_range)*dist_width_pcs.shape[1])
dist_width_score=np.zeros((len(dist_width_range)*dist_width_pcs.shape[1],18))
for i in range(len(dist_width_range)):
    for j in range(dist_width_pcs.shape[1]):
        dist_width[i*dist_width_pcs.shape[1]+j]=dist_width_range[i]
        dist_width_score[i*dist_width_pcs.shape[1]+j,:]=dist_width_pcs[i,j,:18]

prox_width=np.zeros(len(prox_width_range)*prox_width_pcs.shape[1])
prox_width_score=np.zeros((len(prox_width_range)*prox_width_pcs.shape[1],18))
for i in range(len(prox_width_range)):
    for j in range(prox_width_pcs.shape[1]):
        prox_width[i*prox_width_pcs.shape[1]+j]=prox_width_range[i]
        prox_width_score[i*prox_width_pcs.shape[1]+j,:]=prox_width_pcs[i,j,:18]
```

```python
prox_shift_r=np.zeros(18)
dist_strength_r=np.zeros(18)
prox_strength_r=np.zeros(18)
dist_width_r=np.zeros(18)
prox_width_r=np.zeros(18)

print('Proximal shift')
for pc in range(18):
    res=spearmanr(prox_t0,prox_shift_score[:,pc])
    prox_shift_r[pc]=res.correlation
    min_score=np.min(np.mean(prox_shift_pcs[:,:,pc],axis=1))
    max_score=np.max(np.mean(prox_shift_pcs[:,:,pc],axis=1))
    min_percentile=(np.sum(data_pcs[:,pc] < min_score) / len(data_pcs[:,pc]))*100
    max_percentile=(np.sum(data_pcs[:,pc] < max_score) / len(data_pcs[:,pc]))*100
    print('PC%d: r=%0.2f, p=%0.3f, %.2f-%.2f percentile' % ((pc+1), res.correlation, res.pvalue, min_percentile, max_percentile))
    
    
print('\nDistal strength')
for pc in range(18):
    res=spearmanr(dist_ampa,dist_strength_score[:,pc])
    dist_strength_r[pc]=res.correlation
    min_score=np.min(np.mean(dist_strength_pcs[:,:,pc],axis=1))
    max_score=np.max(np.mean(dist_strength_pcs[:,:,pc],axis=1))
    min_percentile=(np.sum(data_pcs[:,pc] < min_score) / len(data_pcs[:,pc]))*100
    max_percentile=(np.sum(data_pcs[:,pc] < max_score) / len(data_pcs[:,pc]))*100
    print('PC%d: r=%0.2f, p=%0.3f, %.2f-%.2f percentile' % ((pc+1), res.correlation, res.pvalue, min_percentile, max_percentile))
    
print('\nProximal strength')
for pc in range(18):
    res=spearmanr(prox_ampa,prox_strength_score[:,pc])
    prox_strength_r[pc]=res.correlation
    min_score=np.min(np.mean(prox_strength_pcs[:,:,pc],axis=1))
    max_score=np.max(np.mean(prox_strength_pcs[:,:,pc],axis=1))
    min_percentile=(np.sum(data_pcs[:,pc] < min_score) / len(data_pcs[:,pc]))*100
    max_percentile=(np.sum(data_pcs[:,pc] < max_score) / len(data_pcs[:,pc]))*100
    print('PC%d: r=%0.2f, p=%0.3f, %.2f-%.2f percentile' % ((pc+1), res.correlation, res.pvalue, min_percentile, max_percentile))
    
print('\nDistal width')
for pc in range(18):
    res=spearmanr(dist_width,dist_width_score[:,pc])
    dist_width_r[pc]=res.correlation
    min_score=np.min(np.mean(dist_width_pcs[:,:,pc],axis=1))
    max_score=np.max(np.mean(dist_width_pcs[:,:,pc],axis=1))
    min_percentile=(np.sum(data_pcs[:,pc] < min_score) / len(data_pcs[:,pc]))*100
    max_percentile=(np.sum(data_pcs[:,pc] < max_score) / len(data_pcs[:,pc]))*100
    print('PC%d: r=%0.2f, p=%0.3f, %.2f-%.2f percentile' % ((pc+1), res.correlation, res.pvalue, min_percentile, max_percentile))
    
print('\nProximal width')
for pc in range(18):
    res=spearmanr(prox_width,prox_width_score[:,pc])
    prox_width_r[pc]=res.correlation
    min_score=np.min(np.mean(prox_width_pcs[:,:,pc],axis=1))
    max_score=np.max(np.mean(prox_width_pcs[:,:,pc],axis=1))
    min_percentile=(np.sum(data_pcs[:,pc] < min_score) / len(data_pcs[:,pc]))*100
    max_percentile=(np.sum(data_pcs[:,pc] < max_score) / len(data_pcs[:,pc]))*100
    print('PC%d: r=%0.2f, p=%0.3f, %.2f-%.2f percentile' % ((pc+1), res.correlation, res.pvalue, min_percentile, max_percentile))
    
```
